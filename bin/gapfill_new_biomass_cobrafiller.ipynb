```python
##for gapfilling purposes, also load in the universal seed model to pool with reactions from model
model = cobra.io.read_sbml_model("../results/reconstructions/psy6_ex.xml")
universal = load_universal_modelseed()
old_model = cobra.io.read_sbml_model('../results/reconstructions/v5_with_all_annotations.xml')
```

```python
cpd16565_c = universal.metabolites.cpd16565_c
cpd16565_e = universal.metabolites.cpd16565_e
rxn1 = Reaction("cor0013_c")
rxn1.name = "Coronatine Ligase"
rxn1.upper_bound = 1000
rxn1.lower_bound = 0
rxn1.add_metabolites({model.metabolites.cpd11406_c:-1,
                     model.metabolites.cor00002_c: -1,
                     cpd16565_c:1})

#create a secreation transport reaction for coronatine
tpt1 = Reaction("cor0014_c")
tpt1.name = ("Coronatine Transport")
tpt1.upper_bound = 1000
tpt1.lower_bound = 0
tpt1.subsystem = "Transport"
tpt1.add_metabolites({cpd16565_c : -1,
                     model.metabolites.cpd16565_e : 1})
model.add_reactions([tpt1, rxn1])
```

```python
for met in model.metabolites:
    if met not in universal.metabolites:
        print (met.id , met.name)
        print (met.reactions)
#phosphotidylethanolamine is making an issue as well, replacing with CDP-ethanolamine which is direclt before adding to DAG moeity     
model.remove_reactions(["pslpdhg_c"] , remove_orphans = True)
cpd11455_c = universal.metabolites.cpd11455_c.copy()
cpd00444_c = universal.metabolites.cpd00444_c.copy()
cpd11652_c = universal.metabolites.cpd11652_c.copy()
cpd11715_c = universal.metabolites.cpd11715_c.copy()
cpd12801_c = universal.metabolites.cpd12801_c.copy()
phospholipid_head = Metabolite(
    'phospholipid_head',
    name = "Phospholipid Head componant of Biomass",
    compartment = "c")

rxn1 = Reaction('pslpdhg_c')
rxn1.name = "Phospholipid Head componant of Biomass"
rxn1.lower_bound = 0
rxn1.upper_bound = 1000
rxn1.add_metabolites({cpd11455_c : -0.2,
                     cpd00444_c: -0.2,
                     cpd11652_c : -0.2,
                     cpd11715_c : -0.2,
                     cpd12801_c : -0.2,
                     phospholipid_head : 1})
model.add_reactions ({rxn1})
```

```python
#model.metabolites.cpd11416_c.remove_from_model()
#model.metabolites.cor00011_c.remove_from_model()
model.metabolites.cpd16565_e.remove_from_model()
cobra.manipulation.delete.prune_unused_metabolites(model)
for met in model.metabolites:
    if met not in universal.metabolites:
        print (met.id , met.name)
        print (met.reactions)
model.repair()
model.objective = "biomass_pto"
```

```python
#add in exchange reactions for all media componants, as well as secreation for all metabolites (upper bound > 0)
for met in model.metabolites:
    if "EX" in str(met.reactions):
        pass
    #add in secreation exchange reactions for internal metabolites
    elif "_c" in met.id:
        rxn = Reaction("EX_" + met.id)
        rxn.lower_bound = 0
        rxn.upper_bound = 1000
        rxn.add_metabolites({met : -1})
        model.add_reactions([rxn])
        print ("I have added exchange reaction " + met.id)
rxn1 = Reaction("EX_phospholipid")
rxn1.lower_bound = 0
rxn1.upper_bound = 1000
rxn1.add_metabolites({model.metabolites.phospholipid : -1})
rxn2 = Reaction("EX_phospholipid_head")
rxn2.lower_bound = 0
rxn2.upper_bound = 1000
rxn2.add_metabolites({model.metabolites.phospholipid_head : -1})
rxn3 = Reaction("EX_fatty_acid")
rxn3.lower_bound = 0
rxn3.upper_bound = 1000
rxn3.add_metabolites({model.metabolites.fatty_acid : -1})
model.add_reactions([rxn1, rxn2, rxn3])       
```

```python
#using the built in cobra gapfiller
from cobra.flux_analysis import gapfill
solution = gapfill(model, universal, lower_bound = 0.05, demand_reactions=True , exchange_reactions=True)
for reaction in solution[0]:
    print(reaction.id)
```
