```python
### purpose: scrape kegg to get the subsytem for each gene
url = "https://www.genome.jp/dbget-bin/www_bget?pst:"


complete_pathway_dict = {}
    
for x in complete_unique:
    opened_url = urllib.request.urlopen(url + x)
    soup = bs4.BeautifulSoup(opened_url)
    table = soup.find("table")
    if table != None:
# The first tr contains the field names.
        headings = [th.get_text().strip() for th in table.find("tr").find_all("th")]
#    datasets = []
        for row in table.find_all("tr")[1:]:
            dataset = dict(zip(headings, (td.get_text() for td in row.find_all("td"))))
#            if "Pathway" in dataset:
            #datasets.append(dataset)
            for key, value in dataset.items():
                if "Pathway" in value:
                    #print (value)
                    result = re.search('(?s)Pathway(.*)Module', value)
                    if result:
                        #print(result.groups())
                        complete_pathway_dict[x] = result.groups()
                    else: pass
                    result1 = re.search('(?s)Pathway(.*)Brite', value)
                    #if result1:
                        #print(result1.groups())
                        #pathway_dict[x] = result.groups()

                else: pass
a = {}

for x, y in complete_pathway_dict.items():
    #print (y)
    for m in y:
        result = re.split('\\xa0\\xa0', m)
        b= []
        for go in result:
            result1 = re.split('pst', go)
            #print ("****")
            b.append(result1[:1])
        print (b)
    a[x] = b
```

```python
#print (a)
complete_go_terms = []
for key, value in a.items():
    for go in value:
        if go in complete_go_terms:
            pass
        else:
            complete_go_terms.append(go)
complete_dict = {}
for go in complete_go_terms:
    #print (go , "=========================")
    go_list = []
    for key, value in a.items():
        if go in value:
            #print (value)
            go_list.append(key)
            #print(go_list)
        else:
            pass
    l = (len(go_list))
    complete_dict[str(go)] = l
complete_dict
```
