```python
mm_ids = [ids for ids in gluc_mm.ids]
for y in mm_ids:
    if "nan":
        pass
    else:
        print (y)
mm_gene_list=[]
for x in mm_ids:
    x = str(x)
    if 'nan' in x:
        pass
    else:
        result = re.split("frozenset\\({'", x)
    for y in result:
        if "frozenset" in y:
            pass
        else:
            mm_gene_list.append(str(y[:-3]))
mm_unique = []
for x in mm_gene_list:
    if x in mm_unique:
        pass
    else:
        mm_unique.append(x)
mm_unique
```

```python
complete_ids = [ids for ids in complete.ids]
for y in complete_ids:
    if "nan":
        pass
    else:
        print (y)
complete_gene_list=[]
for x in complete_ids:
    x = str(x)
    if 'nan' in x:
        pass
    else:
        result = re.split("frozenset\\({'", x)
    for y in result:
        if "frozenset" in y:
            pass
        else:
            complete_gene_list.append(str(y[:-3]))
```

```python
#only get the unique values in the list because of the repeats throughout
complete_unique = []
for x in complete_gene_list:
    if x in complete_unique:
        pass
    else:
        complete_unique.append(x)
complete_unique
```

```python
### purpose: scrape kegg to get the subsytem for each gene
url = "https://www.genome.jp/dbget-bin/www_bget?pst:"


complete_pathway_dict = {}
    
for x in complete_unique:
    opened_url = urllib.request.urlopen(url + x)
    soup = bs4.BeautifulSoup(opened_url)
    table = soup.find("table")
    if table != None:
# The first tr contains the field names.
        headings = [th.get_text().strip() for th in table.find("tr").find_all("th")]
#    datasets = []
        for row in table.find_all("tr")[1:]:
            dataset = dict(zip(headings, (td.get_text() for td in row.find_all("td"))))
#            if "Pathway" in dataset:
            #datasets.append(dataset)
            for key, value in dataset.items():
                if "Pathway" in value:
                    #print (value)
                    result = re.search('(?s)Pathway(.*)Module', value)
                    if result:
                        #print(result.groups())
                        complete_pathway_dict[x] = result.groups()
                    else: pass
                    result1 = re.search('(?s)Pathway(.*)Brite', value)
                    #if result1:
                        #print(result1.groups())
                        #pathway_dict[x] = result.groups()

                else: pass
a = {}

for x, y in complete_pathway_dict.items():
    #print (y)
    for m in y:
        result = re.split('\\xa0\\xa0', m)
        b= []
        for go in result:
            result1 = re.split('pst', go)
            #print ("****")
            b.append(result1[:1])
        print (b)
    a[x] = b
```

```python
#print (a)
complete_go_terms = []
for key, value in a.items():
    for go in value:
        if go in complete_go_terms:
            pass
        else:
            complete_go_terms.append(go)
complete_dict = {}
for go in complete_go_terms:
    #print (go , "=========================")
    go_list = []
    for key, value in a.items():
        if go in value:
            #print (value)
            go_list.append(key)
            #print(go_list)
        else:
            pass
    l = (len(go_list))
    complete_dict[str(go)] = l
complete_dict
```
